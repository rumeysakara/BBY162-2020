```
from google.colab import drive
drive.mount('/gdrive')
def KitapEkle():
  dosya = "/gdrive/My Drive/Colab Notebooks/bby162/veritabani.txt"
  f = open(dosya, "a",encoding="utf-8")
  kitabinAdi = input("Kitap Adı: ")
  yazarAdi = input("Yazar Adı: ")
  basimYili = input("Basım Yılı: ")
  isbn = input("isbn : ")
  yayinEvi = input("Yayın Evi: ")
  f.write(kitabinAdi+"|" + yazarAdi + "|" + basimYili + "|" + isbn + "|" + yayinEvi+"\n")
  print("Kitap Kayıt edildi.")
  f.close()

def KitaplariListele():
  dosya = "/gdrive/My Drive/Colab Notebooks/bby162/veritabani.txt"
  f = open(dosya, "r", encoding='utf-8')
  print(f.read())
  #for line in f.readlines():
  #  print(line)
  f.close()


def KayitAraSilDuzelt():
  dosya = "/gdrive/My Drive/Colab Notebooks/bby162/veritabani.txt"
  f = open(dosya, "r", encoding='utf-8')
  KitapListesi = f.readlines()
  f.close()
  Kayit = input("Eserin adını giriniz: ")
  for kitap in KitapListesi:
    name = kitap.split("|")
    if Kayit == name[0]:
      print("Eser Bilgileri: "+name[0]+","+name[1]+","+name[2]+","+name[3]+","+name[4]+"\n")
      print("Silmek için (s)'ye güncellemek için (g)'ye basınız\n")
      komut =input("İşlem:")
      KitapListesi.remove(kitap)
      f = open(dosya, "w", encoding='utf-8')
      for kitap in KitapListesi:
        f.write(kitap)
      f.write("\n")
      f.close()
      if(komut == "s" or komut == "S"):
        print("Kayıt Silindi")
      elif komut == "g" or komut == "G":
        KitapEkle()
      break

islemTipi = 0
while not islemTipi == 5 :
  print("İşlem Seçiniz")
  print("1)Kitap ekle")
  print("2)Kitapları Listele")
  print("3)Kitap Ara/Sil/Düzelt")
  print("4)Çıkış")
  islemTipi = int (input("Seçilen İşlem:"))
  if islemTipi == 1:
    KitapEkle()
  elif islemTipi == 2:
    KitaplariListele()
  elif islemTipi == 3:
    KayitAraSilDuzelt()
  elif islemTipi == 4:
    break
  else:
    print("Yanlış işlem Seçtiniz")
    continue
```
